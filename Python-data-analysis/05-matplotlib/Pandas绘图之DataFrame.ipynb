```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
# 1-10  取40个数字
df = DataFrame(
    np.random.randint(1,10,40).reshape(10,4),
    columns=['A','B','C','D']
)
df
```

```python
df.plot()
```

```python
df.plot(kind='bar')
plt.show()
```

```python
df.plot(kind='barh')
plt.show()
```

```python
df.plot(kind='bar', stacked=True) #同一个index 不同columns 的数据堆叠
plt.show()
```

```python
df.plot(kind='area')
plt.show()
```
