```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s = Series([1,2,3,4,5])
s
```

```python
s.cumsum() # 累加求和
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind='line',grid=True, label='S1', title='This is Series')
s2.plot(label='S2')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
s1[0:10].plot(ax=ax[0], label='S1', kind='bar') # s1取了10个点
s1[0:10].plot()
s2.plot(ax=ax[1], label='S2')
plt.show()
```
